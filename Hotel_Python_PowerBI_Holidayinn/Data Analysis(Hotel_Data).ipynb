# Data Analysins abd Visualization on clean data

```python
cancelled_perc= df['is_canceled'].value_counts(normalize=True)
cancelled_perc*100
```

```python
#lets visaulaize
plt.figure(figsize=(5,4))
plt.title('reservation status count')
plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts())
plt.show()
```

```python
#lets see from which hotel we are getting maximum cancellations
plt.figure(figsize=(8,4))
ax1= sns.countplot(x='hotel', hue='is_canceled',data=df,palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
#ax1.legend(bbox_to_anchor(1,1))
plt.title('reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.show()
```

```python
#lets check the cancellation percentage of resort and city
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel= df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#we can check the particular date of cancellation
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
#visuallize the data for average price 
plt.figure(figsize=(20,8))
plt.title('Average Daily rate in city and resort',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
#let see in which month and day were the maximum cancellations
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(20,8))
ax1=sns.countplot(x='month',hue='is_canceled', data=df)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not_canceled','canceled'])
plt.show()
```

```python
#we are checking id their is any effect of price on cancellations

plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=20)
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.legend(fontsize=20)
plt.show()

#if we read the graph , then adr , lowest is August when have maxi reservations and have highest cancellanintin in Jan due to high prices
```

```python
#top countries with max cancellations
cancelled_data= df[df['is_canceled']==1] #filter he cancelled data and move to cancelled object
top_10_country=cancelled_data['country'].value_counts()[:10] #fileter the top 10 country , usined values conts give 10 in decreating order and :10 will give and save in series
plt.figure(figsize=(20,8))
plt.title('top 10 countrries with cancellations')
plt.pie(top_10_country, autopct='%.2f',labels=top_10_country.index ) #autopct means upto 2 decimacals
plt.show()
#as per pie, portugal has hihest cancellations almost 70%
```

```python
#let see the cancelled and not cancelled data
cancelled_df_adr= cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data= df[df['is_canceled']==0]
not_cancelled_df_adr= not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average dailty rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
```
